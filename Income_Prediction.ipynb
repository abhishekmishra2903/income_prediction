### Importing the training data. 
### Assigning appropriate field names and constructing a dictionary.
### Converting the dictionary into dataframe.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data=open('adult.data','r')
data_read=data.read()

#splitting the long string into records
data_list=data_read.split('\n')

# Framing a list that contains dictionaries of records.
data_dict=[]

# ignoring the last 2 records because due to some discrepancy in data, those records are empty
for i in range(len(data_list)-2):
    field=data_list[i].split(',')
    dictionary=dict(age=field[0],workclass=field[1],fnlwgt=field[2],education=field[3],education_num=field[4],
                    marital_status=field[5],occupation=field[6],relationship=field[7],race=field[8],sex=field[9],
                    capital_gain=field[10],capital_loss=field[11],hours_per_week=field[12],country=field[13],income=field[14])
    data_dict.append(dictionary)

# Forming the dataframe and printing for visualisation
df_train=pd.DataFrame(data_dict)
df_train
```

### Let's form 2 lists of ages that falls into a particular income category.
### The stacked histogram of the two lists suggests that there isn't a strong trend yet the field can be kept for further analysis.

```python
# Age ranges from 17 to 90. Let's divide it into 5 groups to visualise impace of age on income
# The group be as: under 25, 25-35, 35-50, 50-65, 65+ 

# getting list of age of all poor and rich separately
l_poor=list(df_train['age'][df_train['income']==' <=50K'])
l_rich=list(df_train['age'][df_train['income']==' >50K'])

# converting the values of list from string to numbers
l_poor=[int(i) for i in l_poor]
l_rich=[int(i) for i in l_rich]

# defining a empty list where we will append 5 numbers which will be count of people in each age category, from l_poor list
l_poor_category=[]

#counting people in 17-25 age group and appending in the l_poor_category list.
k=0
for i in range(17,25):
    k=k+l_poor.count(i)
l_poor_category.append(k)

#counting people in 25-35 age group and appending in the l_poor_category list.
k=0
for i in range(25,35):
    k=k+l_poor.count(i)
l_poor_category.append(k)

#counting people in 35-50 age group and appending in the l_poor_category list.
k=0
for i in range(35,50):
    k=k+l_poor.count(i)
l_poor_category.append(k)

#counting people in 50-65 age group and appending in the l_poor_category list.
k=0
for i in range(50,65):
    k=k+l_poor.count(i)
l_poor_category.append(k)

#counting people in 65-90 age group and appending in the l_poor_category list.
k=0
for i in range(65,91):
    k=k+l_poor.count(i)
l_poor_category.append(k)

# defining a empty list where we will append 5 numbers which will be count of people in each age category, from l_rich list
l_rich_category=[]

#counting people in 17-25 age group and appending in the l_rich_category list.
k=0
for i in range(17,25):
    k=k+l_rich.count(i)
l_rich_category.append(k)

#counting people in 25-35 age group and appending in the l_rich_category list.
k=0
for i in range(25,35):
    k=k+l_rich.count(i)
l_rich_category.append(k)

#counting people in 35-50 age group and appending in the l_rich_category list.
k=0
for i in range(35,50):
    k=k+l_rich.count(i)
l_rich_category.append(k)

#counting people in 50-65 age group and appending in the l_rich_category list.
k=0
for i in range(50,65):
    k=k+l_rich.count(i)
l_rich_category.append(k)

#counting people in 65-91 age group and appending in the l_rich_category list.
k=0
for i in range(65,91):
    k=k+l_rich.count(i)
l_rich_category.append(k)

#Defining categories for the base of the plot
categories=['17-25','25-35','35-50','50-65','65-90']

#plotting the two bars in stacked form with one as a base of other
plt.bar(categories,l_rich_category)
plt.bar(categories,l_poor_category,bottom=l_rich_category)
```

### Forming hist plot for workclass.
### There are not >50k people in 'Without-pay' and  'Never-worked' category which is obvious.
### So just for the sake of one on one plotting, removed these two categories from the l1
### There are some interesting trends like in self-employed category everyone is earning above 50K, so we should keep this field for further analysis

```python
# forming two lists for each income category
l1=list(df_train['workclass'][df_train['income']==' <=50K'])
l2=list(df_train['workclass'][df_train['income']==' >50K'])

# removing two categories from l1 as they are absent in l2 and unequal categories will distort stacked histogram.
l1=[i for i in l1 if i not in [' Without-pay',' Never-worked']]

# plotting 
plt.hist(l1, stacked=True, bins=7)
plt.hist(l2, stacked=True, bins=7)
plt.xticks(rotation=90)
plt.show()
```

### This field is related to number of stats collected till the point and is not related to the person concerned, the histogram also confirms the same as distribution of both income group is uniform across the range of fnlwgt.
### So this field can be dropped

```python
# forming two lists for both income categories and converting each entry to numerical
l1=df_train['fnlwgt'][df_train['income']==' <=50K']
l1=[int(i[1:]) for i in l1]
l2=df_train['fnlwgt'][df_train['income']==' >50K']
l2=[int(i[1:]) for i in l2]

# plotting
plt.hist(l1, stacked=True,bins=20)
plt.hist(l2, stacked=True,bins=20)
plt.show()

#dropping the field after visualising the graph
df_train = df_train.drop(['fnlwgt'],axis=1)
```

### The number of unique entries in 'education_num' and 'education' are equal and the field 'education_num'
### seems to be the ordinal transformation of 'education' field. So we can safely drop 'education' field.

```python
print(df_train['education_num'].unique())
print(df_train['education'].unique())

# Dropping education field
df_train = df_train.drop(['education'],axis=1)
```

### As observed from the plot, the income group is fairly depending on the marital status, so we must keep this field.

```python
# forming list of marital_status in both income categories
l1=list(df_train['marital_status'][df_train['income']==' <=50K'])
l2=list(df_train['marital_status'][df_train['income']==' >50K'])

# plotting
plt.hist(l1, stacked=True,bins=7)
plt.hist(l2, stacked=True,bins=7)
plt.xticks(rotation=90)
plt.show()
```

### In some categories e.g. executive managerial, most of the people are in >50k income group. Thus occupation is an important determinant and hence we should keep this field.

```python
# Forming list on occupation for both income categories
l1=list(df_train['occupation'][df_train['income']==' <=50K'])
l2=list(df_train['occupation'][df_train['income']==' >50K'])

# plotting
plt.hist(l1, stacked=True, bins=15)
plt.hist(l2, stacked=True, bins=15)
plt.xticks(rotation=90)
plt.show()
```


### Marital status and relationship are highly correlated fields as can be overved below
### Relationship does give an extra information about gender but that is already accounted in a separate column. So we can safely ignore the relationship field

```python
print(df_train['marital_status'].unique())
print(df_train['relationship'].unique())

# Dropping the relationship field after observing obvious correlation with marital_status
df_train = df_train.drop(['relationship'],axis=1)
```

### As shown below the number of white people constitute ~85 percent of entries so the field is highly skewed.
### Secondly there isn't any deterministic pattern visible among the 4 races
### So we can drop the field.

```python
# forming list of all 'race' entries in both the income categories
l1=list(df_train['race'][df_train['income']==' <=50K'])
l2=list(df_train['race'][df_train['income']==' >50K'])

# printing the number of White entries.It would help us decide if the entries are skewed
print(len(df_train['race'][df_train['race']==' White']))

# plotting
plt.hist(l1, stacked=True)
plt.hist(l2, stacked=True)
plt.xticks(rotation=90)
plt.show()

# dropping the 'race' field after observation
df_train = df_train.drop(['race'],axis=1)
```

### In sex field it is clearly visible that gender is determining income group. So field is important.

```python
# forming lists of entries in 'sex' for both the income groups
l1=list(df_train['sex'][df_train['income']==' <=50K'])
l2=list(df_train['sex'][df_train['income']==' >50K'])

# plotting
plt.hist(l1, stacked=True)
plt.hist(l2, stacked=True)
plt.xticks(rotation=90)
plt.show()
```

### Hours per week of work is a potent determinant of income and it's a numeric value, so keeping it.

### As observed below, 85 percent of data belongs to USA.
### On magnifying and observing, there isn't any obvious trend in income group w.r.t country
### So taking into account the above two factors we can drop the field.

```python
# forming list of county for both the income groups
l1=list(df_train['country'][df_train['income']==' <=50K'])
l2=list(df_train['country'][df_train['income']==' >50K'])

# plotting
plt.hist(l1, stacked=True,bins=42)
plt.hist(l2, stacked=True,bins=42)
plt.xticks(rotation=90)
plt.show()

# printing count of United-States, to decide if the field is imbalanced
print(list(df_train['country']).count(' United-States'))

# dropping the field agter observation
df_train = df_train.drop(['country'],axis=1)
```

### Since in capital_gain and capital_loss fields, 92 percent and 95 percent values are null/0, we should ignore these fields.

```python
# printing the count of null values to decide if the field can be considered for our model or not.
print(list(df_train['capital_gain']).count(' 0'))
print(list(df_train['capital_loss']).count(' 0'))

#dropping both fields as they contain unacceptable number of null values
df_train = df_train.drop(['capital_gain'],axis=1)
df_train = df_train.drop(['capital_loss'],axis=1)

# printing the dataframe for visualisation after all fields are manually inspected
df_train
```

### For the categorical fields i.e. workclass, marital_status and occupation we can go for one hot encoding.

```python
# encoding workclass and then dropping the original categorical field
df_train=pd.concat([df_train,pd.get_dummies(df_train.workclass).add_prefix('workclass_')],axis=1)
df_train=df_train.drop('workclass',axis=1)

# encoding marital_status and then dropping the original categorical field
df_train=pd.concat([df_train,pd.get_dummies(df_train.marital_status).add_prefix('marital_status_')],axis=1)
df_train=df_train.drop('marital_status',axis=1)

# encoding occupation and then dropping the original categorical field
df_train=pd.concat([df_train,pd.get_dummies(df_train.occupation).add_prefix('occupation_')],axis=1)
df_train=df_train.drop('occupation',axis=1)
```

### The fields 'workclass_ ?' and 'occupation_ ?' pertains to no entry in the field, so we can remove these for decluttering

```python
# dropping the fields that got formed by encoding of null entries in workclass and occupation
df_train=df_train.drop('workclass_ ?',axis=1)
df_train=df_train.drop('occupation_ ?',axis=1)
```

```python
df_train
```

### Binary encoding sex & income and converting all Booleans to integer

```python
# applying binary encoding to 'sex' and 'income'
df_train['sex']=df_train['sex'].apply(lambda x:1 if x==' Male' else 0)
df_train['income']=df_train['income'].apply(lambda x:1 if x==' >50K' else 0)

# converting all data to int type
df_train=df_train.astype(int)
df_train
```

### Importing test data

```python
# loading test file
data=open('adult.test','r')
data_read=data.read()

# splitting records
data_list=data_read.split("\n")

# popping 1st record as it was some kind of label
data_list.pop(0)

# framing a list of dictionaries of all records
data_dict=[]
for i in range(len(data_list)-2):
    field=data_list[i].split(',')
    dictionary=dict(age=field[0],workclass=field[1],fnlwgt=field[2],education=field[3],education_num=field[4],
                    marital_status=field[5],occupation=field[6],relationship=field[7],race=field[8],sex=field[9],
                    capital_gain=field[10],capital_loss=field[11],hours_per_week=field[12],country=field[13],income=field[14])
    data_dict.append(dictionary)

# forming dataframe
df_test=pd.DataFrame(data_dict)
df_test
```

### Applying exactly same transformations to the test data

```python
# repeating the same data cleaning steps 
df_test = df_test.drop(['fnlwgt'],axis=1)
df_test = df_test.drop(['education'],axis=1)
df_test = df_test.drop(['relationship'],axis=1)
df_test = df_test.drop(['race'],axis=1)
df_test = df_test.drop(['country'],axis=1)
df_test = df_test.drop(['capital_gain'],axis=1)
df_test = df_test.drop(['capital_loss'],axis=1)

df_test=pd.concat([df_test,pd.get_dummies(df_test.workclass).add_prefix('workclass_')],axis=1)
df_test=df_test.drop('workclass',axis=1)

df_test=pd.concat([df_test,pd.get_dummies(df_test.marital_status).add_prefix('marital_status_')],axis=1)
df_test=df_test.drop('marital_status',axis=1)

df_test=pd.concat([df_test,pd.get_dummies(df_test.occupation).add_prefix('occupation_')],axis=1)
df_test=df_test.drop('occupation',axis=1)

df_test=df_test.drop('workclass_ ?',axis=1)
df_test=df_test.drop('occupation_ ?',axis=1)

df_test['sex']=df_test['sex'].apply(lambda x:1 if x==' Male' else 0)
df_test['income']=df_test['income'].apply(lambda x:1 if x==' >50K.' else 0)
df_test=df_test.astype(int)
df_test
```

### Defining X and Y axis for both training and testing data

```python
train_Y=df_train['income']
train_X=df_train.drop('income',axis=1)

test_Y=df_test['income']
test_X=df_test.drop('income',axis=1)
```

### In our data, out of 33 only 3 columns are numerical, rest are binary. So, taking decision tree as the first choice for the classification.
### For the Classifier below, on hit and trial, got optimum result at max_depth=10

```python
from sklearn import tree

classifier = tree.DecisionTreeClassifier(max_depth=10)
classifier = classifier.fit(train_X, train_Y)

print(classifier.score(train_X,train_Y))
print(classifier.score(test_X,test_Y))
```

### Let's print the confusion matrix to get a broad idea. Let's check for the report in next step.

```python
from sklearn.metrics import classification_report, confusion_matrix

pred_Y = classifier.predict(test_X)
result = confusion_matrix(test_Y,pred_Y)
print(result)
```

### In the report below, everything looks fine for 0 but scores of 1 are on the lower side. One of the reasons can be that the dataset is skewed in favour of 0. There are only ~16 percent data pertaining to <=50K salary group.

```python
result1 = classification_report(test_Y,pred_Y)
print (result1)
```

### Let's try another classifier i.e. random forest.

###  I tried adjusting some parameters like n_estimators, max_features, max_depth, max_samples etc. Some significant improvement was observed with max_depth=10, so keeping that.

```python
from sklearn.ensemble import RandomForestClassifier

forest=RandomForestClassifier(max_depth=10)
forest.fit(train_X,train_Y)
print(forest.score(train_X,train_Y))
print(forest.score(test_X,test_Y))
```

### Let's print confusion matrix to visualize the distribution of score across both the categories.

```python
pred_Y = forest.predict(test_X)
result = confusion_matrix(test_Y,pred_Y)
print(result)
```

### As expected from random forest, the report parameters have slightly improved than the previous model

```python
result1 = classification_report(test_Y,pred_Y)
print (result1)
```
